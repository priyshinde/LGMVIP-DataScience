### *IMAGE TO PENCIL SKETCH WITH PYTHON*

**Step-1:  Importing required libraries**

Letâ€™s first start by importing modules into our program. We will be making use of the OpenCV functions and matplotlib module to plot the images.

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

```python
import  matplotlib.pyplot as plt
```

**Step-2:  Loading the image**

we will be loading the image file into our program by making use of the imread function which takes the path of the file.

```python
img = cv2.imread("/content/Screenshot 2022-01-08 195643.jpg")  #  imread()- This function will load the image i.e in the specified folder. 


#Displaying the original image 
plt.figure(figsize=(10,20))
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

**Step-3: Converting the image into an gray scale image**

To decrease the complexity of the image and make the handling easier we will be converting the image to a grayscale image with the help of the cvtColor function.

```python
#Converting Image to Gray Scale
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #cvtColor()- This function takes color as an argument and then changes the source image color into that color.
plt.figure(figsize=(10,20))
plt.title("Gray Image")
plt.imshow(gray_img,cmap='gray')
plt.axis("off")
plt.show()
```

**Step-4: Inverting the gray image**

Now the next step is to invert the image. Now your question would be why to do that? The answer to the same is that when a image is inverted it helps in studying the image with more precision and detail.

The more detailed study is important for the sketch generation as the sketches need to be accurate and good. 

```python
#Inverting the Gray Image
inverted_gray_img=cv2.bitwise_not(gray_img)
plt.figure(figsize=(10,20))
plt.title("Inverted Gray Image")
plt.imshow(inverted_gray_img,cmap='gray')
plt.axis("off")
plt.show()
```

**Step-5: Blurring the image**

We have used the gaussian blur technique with 21 x 21 pixel and the default sigma values filter on the image to smoothen our image.It is used to reduce the noise in the image.

```python
#Blur the image using Gaussian function
blur_img=cv2.GaussianBlur(inverted_gray_img,(21,21),sigmaX=0, sigmaY=0)  #GaussianBlur()- This function is used to modify the image by sharpening the edges of the image, smoothen the image.
plt.figure(figsize=(10,20))
plt.title("Blur Image")
plt.imshow(blur_img,cmap='gray')
plt.axis("off")
plt.show()
```

**Step-6: Obtaining and displaying the final sketch**

```python
#Final Sketch Image
pencil_sketch = cv2.divide(gray_img,255-blur_img,scale = 256) #By using this function, it is dividing the greyscale value of the image by the inverse of blurred image value which highlights the boldest edges.
plt.figure(figsize=(10,15))
plt.title("Final Sketch Image")
plt.imshow(pencil_sketch,cmap = 'gray')
plt.axis("off")
plt.show()
```
